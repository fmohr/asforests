```python
import json
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import os
```

```python
def get_performance_curves(openmlid, folder="../experiments/empirical_curves/results"):

    folder = f"{folder}/{openmlid}"
    
    result_files = sorted([f for f in os.listdir(folder) if f.endswith(".json")])
    data = []
    cols = []
    for name in result_files:
        with open(f"{folder}/{name}", "r") as f:
            curve = json.load(f)
            data.append(curve["scores"])
            cols.append(name)

    # Find the maximum length
    max_len = max(len(lst) for lst in data)
    
    # Pad each list with NaNs
    padded_data = [lst + [np.nan] * (max_len - len(lst)) for lst in data]
    
    return pd.DataFrame(padded_data, columns=range(1, max_len + 1), index=cols)
```
