```python
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
```

```python
oob_probability = 0.368 # this is the probability of "success", i.e. that a specific instance is OOB for a tree
```

```python
t = 100

print(f"Mean value of trees used on each instance: {np.round(t * oob_probability, 1)}")

domain = np.arange(0, t)

fig, ax = plt.subplots()
ax.plot(domain, scipy.stats.binom.pmf(k = domain, n = t, p = oob_probability))
plt.show()
```
